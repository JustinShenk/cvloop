```python
# %matplotlib nbagg
from cvloop import cvloop
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import cv2
import sys
import random

FRAME_WIDTH = 640.
FRAME_HEIGHT = 480.
cascPath = "examples/face.xml"
faceCascade = cv2.CascadeClassifier(cascPath)

def findFaces(frame):
    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale( \
        frame_gray, \
        scaleFactor=1.3, \
        minNeighbors=5, \
        minSize=(50, 50), \
        #         flags=cv2.cv.CV_HAAR_SCALE_IMAGE
        flags=0)
    return faces

def getHat():
    hats = ['brownHat.png','xmasHat.png']
    path = 'hat_matchine/'
    randomHat = path + random.choice(hats)
    _hat = cv2.imread(randomHat,-1)
    b,g,r,a = cv2.split(_hat)
    _hat = cv2.merge((r,g,b,a))
    return _hat

_hat = getHat()

def addHat(frame):
    hat = _hat.copy()
    faces = findFaces(frame)
    xOffset = -20    # Number of pixels left to move hat.
    yOffset = 30    # Number of pixels up to move hat.
    # Extra offset for width and height scaling.
#     wOffset = 80
#     hOffset = 40
    origHatHeight = hat.shape[0]
    origHatWidth = hat.shape[1]
    hatWidth = hat.shape[1]
    for x,y,w,h in faces:
        # Scale hat to fit face.
        hatWidth = w
        hatHeight = hatWidth * origHatHeight // origHatWidth
        hat = cv2.resize(hat, (hatWidth, hatWidth), interpolation=cv2.INTER_AREA)

        # Adjust position of hat in frame with respect to face.
        # y: Face rectangle top-left corner.
        # y0: Top y-position of hat in frame.
        # yOffset: Number of pixels to move hat up.

        # Allow clipping.
        # TODO: Complete this section.
        hatx0 = haty0 = 0
        haty1 = hat.shape[0]
        hatx1 = hat.shape[1]
        
        y0 = y - hat.shape[0] + yOffset
        if y0 < 0: # If hat starts above the frame, clip it.
            haty0 = abs(y0) # Find beginning of hat ROI.
            y0 = 0
        y1 = y0 + hat.shape[0] - haty0
        if y1 > FRAME_HEIGHT:
            haty1 = y1 - FRAME_HEIGHT
            y1 = FRAME_HEIGHT
        x0 = x  - hat.shape[1] + xOffset
        if x0 < 0:
            hatx0 = abs(x0)
            x0 = 0
        x1 = x0 + hat.shape[1] - hatx0
        if x1 > FRAME_WIDTH:
            hatx1 = x1 - FRAME_WIDTH
            x1 = FRAME_WIDTH
        
        for c in range(0, 3):        
            hatSlice = hat[:, :, c] * (hat[:, :, 3] / 255.0)
            backgroundSlice = frame[y0:y1, x0:x1, c] * (1.0 - hat[haty0:haty1, hatx0:hatx1, 3] / 255.0)
            frame[y0:y1, x0:x1, c] = hatSlice + backgroundSlice
        
    return frame

capture = cv2.VideoCapture(0)
capture.set(cv2.CAP_PROP_FRAME_WIDTH, FRAME_WIDTH)
capture.set(cv2.CAP_PROP_FRAME_HEIGHT, FRAME_HEIGHT)

plt, anim = cvloop(source=capture,function=addHat)
plt.show()
```
